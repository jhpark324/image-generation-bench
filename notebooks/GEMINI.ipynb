```python
import sys
sys.path.append('..')

from prompt import PROMPT
from PIL import Image

prompt = PROMPT
image = Image.open("../base_image.png")
```

## nano banana

```python
from google import genai
from google.genai import types

client = genai.Client()

response = client.models.generate_content(
    model="gemini-2.5-flash-image",
    contents=[prompt, image],
)
```

```python
for part in response.parts:
    if part.text is not None:
        print(part.text)
    elif part.inline_data is not None:
        image = part.as_image()
        image.save("../images/nanoBanana_image.png")
```

## nano banana pro

```python
from PIL import Image
img = Image.open('../base_image.png')
img = img.resize((1024, 1536))  # 크기 축소
```

```python
from google import genai
from google.genai import types

aspect_ratio = "9:16" # "1:1","2:3","3:2","3:4","4:3","4:5","5:4","9:16","16:9","21:9"
resolution = "1K" # "1K", "2K", "4K"

client = genai.Client()

response_pro = client.models.generate_content(
    model="gemini-3-pro-image-preview",
    contents=[
        prompt,
        img
    ],
    config=types.GenerateContentConfig(
        response_modalities=['TEXT', 'IMAGE'],
        image_config=types.ImageConfig(
            aspect_ratio=aspect_ratio,
            #image_size=resolution
        ),
    )
)
```

```python
for part in response_pro.parts:
    if part.text is not None:
        print(part.text)
    elif part.inline_data is not None:
        image = part.as_image()
        image.save("../images/nanoBanana_pro_image.png")
```
